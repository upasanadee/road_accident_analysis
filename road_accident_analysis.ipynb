```python
import pandas as pd

# Load the Excel file
excel_file = "/content/extracted_data/Road Accident Data.xlsx"

# Load the first sheet
df = pd.read_excel(excel_file, engine='openpyxl')

# Display dataset information
print(df.info())  # Check dataset structure
print(df.head())  # Preview first few rows

# This code loads an Excel file, reads the first sheet into a Pandas DataFrame, and displays its structure and first few rows.
```

```python
print(f"Total number of accidents: {df.shape[0]}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Yearly distribution
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x="Year", palette="Blues")
plt.title("Accidents Per Year")
plt.xticks(rotation=45)
plt.show()

# Monthly distribution
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x="Month", order=["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"], palette="coolwarm")
plt.title("Accidents Per Month")
plt.xticks(rotation=45)
plt.show()

# Accidents by Day of the Week
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x="Day_of_Week", order=["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"], palette="viridis")
plt.title("Accidents by Day of the Week")
plt.xticks(rotation=45)
plt.show()

# This code visualizes accident data by year, month, and day of the week using Seaborn bar plots with different color palettes.
# It ensures proper label ordering for months and weekdays.
# The plots help identify trends and seasonal variations in accidents.
# It also improves data interpretation by highlighting high-risk periods.
```

```python
# Convert Time to hour format
df["Hour"] = pd.to_datetime(df["Time"], format="%H:%M:%S", errors="coerce").dt.hour

# Plot
plt.figure(figsize=(12, 5))
sns.histplot(df["Hour"].dropna(), bins=24, kde=True, color="red")
plt.title("Accidents by Hour of the Day")
plt.xlabel("Hour of the Day")
plt.ylabel("Number of Accidents")
plt.xticks(range(0, 24))
plt.show()

# This code extracts the hour from the "Time" column and handles errors gracefully.
# It then creates a histogram to visualize the distribution of accidents by hour.
# The kde=True option adds a smooth density curve for better trend analysis.
# The plot helps identify peak accident hours, aiding in traffic safety measures.
```

```python
top_locations = df["Local_Authority_(District)"].value_counts().head(10)

plt.figure(figsize=(12, 5))
sns.barplot(x=top_locations.index, y=top_locations.values, palette="coolwarm")
plt.xticks(rotation=90)
plt.title("Top 10 Locations with Highest Accidents")
plt.ylabel("Number of Accidents")
plt.show()

# This code identifies the top 10 districts with the highest number of accidents.
# It then creates a bar plot to visualize accident distribution across these locations.
# The rotation=90 ensures labels are readable for long district names.
# This helps pinpoint high-risk areas for targeted safety interventions.
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df["Longitude"], y=df["Latitude"], alpha=0.3)
plt.title("Geographical Distribution of Accidents")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()

# This code creates a scatter plot to visualize the geographical distribution of accidents.
# Longitude and latitude are used to map accident locations.
# The alpha=0.3 makes overlapping points more transparent for better clarity.
# This helps identify accident hotspots and spatial patterns.
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="Accident_Severity", palette="pastel")
plt.title("Distribution of Accident Severities")
plt.show()
```

```python
severity_counts = df["Accident_Severity"].value_counts()
fatal_percentage = (severity_counts["Fatal"] / df.shape[0]) * 100

print(f"Percentage of Fatal Accidents: {fatal_percentage:.2f}%")
```
